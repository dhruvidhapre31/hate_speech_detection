# Data Preprocessing

A stop word is a commonly used word (such as “the”, “a”, “an”, “in”) that provides less or no knowledge about the nature of tweet. Hence, stopwords are removed with the help of nltk.

Stemming is an attempt to reduce a word to its stem or root form. Hence, stemming is done on the tweets.

The alphabets are kept while numbers, punctuations and extra spaces are removed from the tweets. 

```python
STOPWORDS = set(stopwords.words('english'))
STOPWORDS.add("rt")
stemmer = nltk.SnowballStemmer("english")

def remove_stopwords(text):
    return ' '.join([word for word in str(text).split() if word not in STOPWORDS])

def stemming(text):
    return ' '.join([stemmer.stem(word) for word in str(text).split()])
    
def text_preprocessing(sen):
    sen = str(sen).lower()
    sen = re.sub('[^a-zA-Z]', ' ', sen)  # remove punctuations and numbers
    sen = re.sub(r'\s+', ' ', sen) # remove extra spaces from the data
    sen = remove_stopwords(sen) # remove stopwords
    sen = stemming(sen) # apply stemming
    return sen
```

```python
data['tweet'] = data['tweet'].apply(text_preprocessing)
data.head()
```

# Splitting the data

```python
x = np.array(data["tweet"])
y = np.array(data["label"])
```

TFIDF (Term Frequency - Inverse Data Frequecny)

The term frequency is the number of occurrences of a specific term in a document.

Document frequency is the number of documents containing a specific term. Document frequency indicates how common the term is.

```python
tfidf_vec = TfidfVectorizer(max_features=1000, min_df=0.01, max_df=0.8)  
# max_features - maximum used words, ignore the rare words
# min_df = words with < 1% occurance (remove)
# max_df = words occuring > 80% (remove)
X_tfidf = tfidf_vec.fit_transform(x).toarray()
X_train_tfidf, X_test_tfidf, y_train_tfidf, y_test_tfidf = train_test_split(X_tfidf, y, test_size=0.3, random_state=100)
```

```python
dt_tfidf = DecisionTreeClassifier()
dt_tfidf.fit(X_train_tfidf, y_train_tfidf)
y_pred_tfidf = dt_tfidf.predict(X_test_tfidf)
```

```python
print(accuracy_score(y_test_tfidf, y_pred_tfidf))
```

```python
print(classification_report(y_test_tfidf, y_pred_tfidf))
```

Count Vectorizer

CountVectorizer creates a matrix in which each unique word is represented by a column of the matrix, and each text sample from the document is a row in the matrix. The value of each cell is nothing but the count of the word in that particular text sample.

```python
cv = CountVectorizer()
X_cv = cv.fit_transform(x).toarray() # Fit the Data
X_train_cv, X_test_cv, y_train_cv, y_test_cv = train_test_split(X_cv, y, test_size=0.3, random_state=100)
```

```python
dt_cv = DecisionTreeClassifier()
dt_cv.fit(X_train_cv, y_train_cv)
y_pred_cv = dt_cv.predict(X_test_cv)
```

```python
print(accuracy_score(y_test_cv, y_pred_cv))
```

```python
print(classification_report(y_test_cv, y_pred_cv))
```

As the accuracy of the count vectorizer technique is higher than that of TFIDF technique, hence we use the model trained with the help of count vectorizer
